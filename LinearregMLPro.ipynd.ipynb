```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

Lets load the Boston House Pricing Dataset

```python
from sklearn.datasets import load_diabetes
diabetes_ds=load_diabetes()
```

Preparing The Dataset

```python
dataset = pd.DataFrame(diabetes_ds.data, columns=diabetes_ds.feature_names)
dataset['target'] = diabetes_ds.target
```

```python
dataset
```

```python
print(dataset.head())
print(dataset.tail())
```

```python
#Checking null or missing values
dataset.isnull().sum()
```

Analysing the Dataset

```python
#  Exploratory Data Analysis
## Correlation

dataset.corr()
```

```python
plt.scatter(dataset['s3'], dataset['sex'])
plt.xlabel("s3")
plt.ylabel("sex")
```

```python
## Independent and Dependent feature's

X = dataset.iloc[ : , :-1]
Y = dataset.iloc[ : , -1:]
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
# Standardize the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit_transform(x_train)
scaler.transform(x_test)
```

Model Training

```python
from sklearn.linear_model import LinearRegression
regression= LinearRegression()
regression.fit(x_train,y_train)
```

```python
## print the coefficients and the intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
## on which parameters, the model has been trained
regression.get_params()
```

```python
## Prediction with Test Data

reg_pred = regression.predict(x_test)
```

```python
reg_pred
```

```python
## plot a scatter plot for the prediction
plt.scatter(y_test,reg_pred)
plt.xlabel('Actual Points')
plt.ylabel('Predicted Points')
plt.title("Linear Regression Model")
```

```python
from sklearn.base import accuracy_score


accuracy_score(reg_pred,y_test)
```

```python
# Residualus means errors
residuals = y_test - reg_pred
```

```python
residuals
```

```python
## scatter plot with respect to prediction and residuals
plt.scatter(reg_pred,residuals)
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Scatter Plot")
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error


print(mean_absolute_error(y_test,reg_pred))
print(mean_squared_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
print(score)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, reg_pred)
print(f"R-squared: {score}")

# Calculate Adjusted R-squared manually
n = x_test.shape[0]  # Number of observations
k = x_test.shape[1]  # Number of independent variables

adjusted_r2 = 1 - (1 - score) * (n - 1) / (n - k - 1)
print(f"Adjusted R-squared: {adjusted_r2}")
```
